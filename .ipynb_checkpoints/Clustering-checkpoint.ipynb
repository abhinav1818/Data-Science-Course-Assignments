```python
import pandas as pd
# Load data
data_full = pd.read_excel('EastWestAirlines.xlsx', sheet_name='data')
print("Original shape:", data_full.shape)
print("Missing values:", data_full.isna().sum().sum())

# Drop ID column
data = data_full.drop(columns=['ID#'])

# Remove outliers beyond 99th percentile in key columns
thresholds = {
    'Balance': data['Balance'].quantile(0.99),
    'Qual_miles': data['Qual_miles'].quantile(0.99),
    'Bonus_miles': data['Bonus_miles'].quantile(0.99),
    'Flight_miles_12mo': data['Flight_miles_12mo'].quantile(0.99)
}
mask = (
    (data['Balance'] <= thresholds['Balance']) &
    (data['Qual_miles'] <= thresholds['Qual_miles']) &
    (data['Bonus_miles'] <= thresholds['Bonus_miles']) &
    (data['Flight_miles_12mo'] <= thresholds['Flight_miles_12mo'])
)
data_clean = data.loc[mask].reset_index(drop=True)
print("Shape after outlier removal:", data_clean.shape)

# Standardize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_clean)
```

```python
#K-Means clustering for different no of cluster

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sse = []
sil_scores = []
ks = range(2, 11)
for k in ks:
    km = KMeans(n_clusters=k, random_state=0, n_init=10, max_iter=200)
    labels = km.fit_predict(data_scaled)
    sse.append(km.inertia_)
    sil_scores.append(silhouette_score(data_scaled, labels))

for k, s, sil in zip(ks, sse, sil_scores):
    print(f"K={k}: SSE={s:.0f}, Silhouette={sil:.3f}")
```

```python
#Visualization elbow method of k-means clustering

plt.figure(figsize=(6,4))
plt.plot(ks, sse, marker='o', color='orange')
plt.title("Elbow Method for K-Means")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (SSE)")
plt.grid(True)
plt.savefig('elbow.png')
```

```python
# Final K-Means with k=3
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
data_clean['cluster_km'] = kmeans.fit_predict(data_scaled)
print("Cluster sizes (KMeans):", data_clean['cluster_km'].value_counts().to_dict())
print("Silhouette (KMeans, k=3):",
      silhouette_score(data_scaled, data_clean['cluster_km']))
```

```python
#selecting best hierarchical linkage method
import numpy as np


hc_ward = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels_hc3 = hc_ward.fit_predict(data_scaled)
print("Cluster sizes (Hierarchical Ward, 3):",
      pd.Series(labels_hc3).value_counts().to_dict())
sil = silhouette_score(data_scaled, labels_hc3)
print(np.round(sil,2))
```

```python
#DBSCAN Method applying with eps = [0.5,1.0,1.5]

from sklearn.cluster import DBSCAN

for eps in [0.5, 1.0, 1.5]:
    db = DBSCAN(eps=eps, min_samples=5)
    labels_db = db.fit_predict(data_scaled)
    n_clusters = len(set(labels_db)) - (1 if -1 in labels_db else 0)
    n_noise = list(labels_db).count(-1)
    if n_clusters > 1:
        sil = silhouette_score(data_scaled[labels_db != -1], labels_db[labels_db != -1])
    else:
        sil = None
    print(f"eps={eps}: clusters={n_clusters}, noise={n_noise}, silhouette={sil:.3f}")
```

```python
#Best DBSCAN method using eps parameter

db = DBSCAN(eps=1.0, min_samples=5)
labels_dbscan = db.fit_predict(data_scaled)
data_clean['cluster_db'] = labels_dbscan
print("DBSCAN cluster sizes:", pd.Series(labels_dbscan).value_counts().sort_index().to_dict())
print("Silhouette (DBSCAN, eps=1):",
      silhouette_score(data_scaled[labels_dbscan!=-1], labels_dbscan[labels_dbscan!=-1]))

#for DBSCAN eps = 1.0 gives best Silhoutte score with 7 clusters and 5 core points
```

```python
#Converting data into 2-D data using PCA
#Visualization of how clusters are formed using K-means

from sklearn.decomposition import PCA

# Project data to 2 principal components
pca = PCA(n_components=3, random_state=0)
coords = pca.fit_transform(data_scaled)

plt.figure(figsize=(6,5))
plt.scatter(coords[:,0], coords[:,1], c=data_clean['cluster_km'], cmap='viridis', s=10 )
plt.title("K-Means Clusters (k=3) in PCA Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label="K-Means Cluster")
plt.savefig('kmeans_scatter.png')
```

```python
#Hierarchical Clustering method
#Visualization of how clusters are formed using Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering

fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes = axes.flatten()

linkages = ['single', 'complete', 'average', 'ward']
for i, method in enumerate(linkages):
    hc = AgglomerativeClustering(n_clusters=3, linkage=method)
    labels_hc = hc.fit_predict(data_scaled)
    sizes = pd.Series(labels_hc).value_counts().sort_index().to_dict()
    sil = silhouette_score(data_scaled, labels_hc) if len(set(labels_hc))>1 else None
    axes[i].scatter(coords[:, 0], coords[:, 1], c=labels_hc, cmap='viridis', s=10)
    axes[i].set_title(f"Hierarchical Clustering ({method.title()} Linkage)")
    axes[i].set_xlabel("PC1")
    axes[i].set_ylabel("PC2")
    print(f"{method.title()} Linkage: sizes={sizes}, Silhouette={sil:.3f}")
    
plt.tight_layout()
plt.show()
```
