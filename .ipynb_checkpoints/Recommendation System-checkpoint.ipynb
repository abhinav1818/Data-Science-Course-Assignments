```python
import pandas as pd

df = pd.read_csv('anime.csv')
df_clean = df.dropna(subset=['genre','type','rating'])        # drop missing genre/type/rating
df_clean = df_clean[df_clean['episodes'] != 'Unknown']       # remove unknown episodes
df_clean['episodes'] = pd.to_numeric(df_clean['episodes'])   # convert to numeric
df_clean = df_clean.dropna(subset=['episodes']).reset_index(drop=True)

print("Cleaned data shape:", df_clean.shape)
df_clean.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distributions
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Anime types
type_counts = df_clean['type'].value_counts()
sns.barplot(x=type_counts.index, y=type_counts.values, ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Anime Types')
axs[0, 0].set_ylabel('Count')
axs[0, 0].set_xticklabels(axs[0, 0].get_xticklabels(), rotation=45)

# Rating distribution
sns.histplot(df_clean['rating'], bins=20, kde=True, ax=axs[0, 1], color='skyblue')
axs[0, 1].set_title('Distribution of Ratings')

# Episode counts
sns.histplot(df_clean['episodes'], bins=50, ax=axs[1, 0],  color='salmon')
axs[1, 0].set_title('Distribution of Episode Counts (Log Scale)')

# Top genres
from collections import Counter
import itertools

genres_split = df_clean['genre'].str.split(', ')
all_genres = list(itertools.chain.from_iterable(genres_split))
genre_counts = Counter(all_genres)
top_genres = genre_counts.most_common(10)
genre_names, genre_values = zip(*top_genres)

sns.barplot(x=list(genre_names), y=list(genre_values), ax=axs[1, 1])
axs[1, 1].set_title('Top 10 Genres')
axs[1, 1].set_xticklabels(axs[1, 1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Data-Preprocessing

from sklearn.preprocessing import MultiLabelBinarizer, MinMaxScaler

# One-hot encode genres (multi-label) and type
genres_list = df_clean['genre'].str.split(', ')
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(genres_list)          # shape (11832, num_genres)
type_dummies = pd.get_dummies(df_clean['type'])        # shape (11832, num_types)

# Scale numerical features
scaler = MinMaxScaler()
num_features = scaler.fit_transform(df_clean[['episodes','rating','members']])

# Combine all features into a single feature matrix
import numpy as np
features = np.hstack([genre_encoded, type_dummies.values, num_features])
print("Combined feature matrix shape:", features.shape)
```

```python
#Data Partition

from sklearn.model_selection import train_test_split

# Split into training and test sets
train_df, test_df, train_features, test_features = train_test_split(
    df_clean, features, test_size=0.2, random_state=42)

# Reset index to align DataFrame rows with feature matrix indices
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

```

```python
#Recommendation Function

from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar(anime_name, features_df, feature_matrix, sim_threshold):
    matches = features_df[features_df['name'].str.lower() == anime_name.lower()]
    if matches.empty:
        raise ValueError(f"Anime '{anime_name}' not found in dataset.")
    idx = matches.index[0]  # use first match
    target_vec = feature_matrix[idx].reshape(1, -1)
    sims = cosine_similarity(target_vec, feature_matrix)[0]
    sims[idx] = -1  # exclude itself
    indices = [i for i, sim in enumerate(sims) if sim >= sim_threshold]
    recommendations = sorted([(features_df.iloc[i]['name'], sims[i]) for i in indices], key=lambda x: x[1], reverse=True)
    return recommendations

```

```python
# Example recommendations for "Death Note"
recs = recommend_similar('Death Note', train_df, train_features, sim_threshold=0.75)
print("Recommendations for 'Death Note':")
for name, score in recs:
    print(f"  {name} (score: {score:.3f})")
```

```python
# Get all anime with similarity ≥ 0.85 to 'One Piece'
recommendations = recommend_similar('Death Note', train_df, train_features, sim_threshold=0.85)

# Print results
print(f"Recommendations for 'Death Note' with similarity ≥ 0.85:")
for name, score in recommendations:
    print(f"  {name} (score: {score:.3f})")
```

```python
#Evalution Metrics

# Prepare genre indicator arrays for train and test
train_genres = train_features[:, :genre_encoded.shape[1]]  # first 43 dims are genre
test_genres = test_features[:, :genre_encoded.shape[1]]


# Evaluation Function for Threshold-Based Filtering in Anime Recommender
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def evaluate_threshold_filtering(train_genres, train_features, train_df, test_genres, test_features, sim_threshold=0.7):
    precisions, recalls, f1s = [], [], []

    for i in range(test_features.shape[0]):
        # Define relevance as sharing at least one genre
        relevant_mask = (train_genres.dot(test_genres[i]) >= 1)
        relevant_indices = np.where(relevant_mask)[0]
        if len(relevant_indices) == 0:
            continue

        # Compute cosine similarities
        sims = cosine_similarity(test_features[i].reshape(1, -1), train_features)[0]
        sims_filtered = [(j, sim) for j, sim in enumerate(sims) if sim >= sim_threshold]
        if not sims_filtered:
            continue

        top_idxs = [j for j, _ in sims_filtered]
        tp = np.isin(top_idxs, relevant_indices).sum()

        prec = tp / len(top_idxs)
        rec = tp / len(relevant_indices)
        f1 = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0

        precisions.append(prec)
        recalls.append(rec)
        f1s.append(f1)

    return np.mean(precisions), np.mean(recalls), np.mean(f1s), len(precisions)

p, r, f1, count = evaluate_threshold_filtering(train_genres, train_features, train_df,
                                               test_genres, test_features, sim_threshold=0.75)
print(f"Evaluated on {count} test anime.")
print(f"Precision: {p:.3f}, Recall: {r:.3f}, F1-score: {f1:.3f}")
```

```python
#Performance of recommendation system

#Precision: 1.000
#Every anime recommended by the system shared at least one genre with the target anime.

#Recall: 0.102 (Low)
#Even though it recommends genre-matched anime, it misses most other potentially relevant titles that share genres but fall below the similarity threshold.

#F1-score: 0.138
#The F1-score is the harmonic mean of precision and recall.
#A high precision with low recall suggests a conservative system — it recommends only the most similar anime and avoids false positives but sacrifices coverage.

```
