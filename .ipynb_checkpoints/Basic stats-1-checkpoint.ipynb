```python
#Descriptive Analytics for Numerical Columns
# 	There are three types of descriptive statistics:
# 1.	Central Tendency: - Mean, Mode, Median
# 2.	Spread: - Range, Standard Deviation, Variation
# 3.	Shape of the Data: - Histogram, Bar plot, Box plot, Scatter plot
```

```python
#step 1 - Collecting the data
import pandas as pd
import numpy as np

df = pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
#Step 2: Identifying the numerical columns
num_columns = df.select_dtypes(include='number')
print(num_columns)
```

```python
#Step 3: Performing the Statistical Measures on Numerical columns (Volume, Avg Price, Total sales value, Discount rate, Discount amount, Net sales value): -

#Volume- column

print("Volume Mean :", df['Volume'].mean())
print("Volume Standard deviation :", df['Volume'].std())
print("Volume Median:", (df['Volume']).median())
print("Volume Mode:",df['Volume'].mode())
print("===========================================================")
#Avg Price- Column

print("Avg Price Mean :", df['Avg Price'].mean())
print("Avg Price Standard deviation :", df['Avg Price'].std())
print("Avg Price Median:", (df['Avg Price']).median())
print("Avg Price Mode:",df['Avg Price'].mode())
print("===========================================================")

#Total Sales Value - column

print("Total Sales Value :", df['Total Sales Value'].mean())
print("Total Sales Value Standard deviation :", df['Total Sales Value'].std())
print("Total Sales Value Median:", (df['Total Sales Value']).median())
print("Total Sales Value Mode:",df['Total Sales Value'].mode())
print("============================================================")

#Discount Rate (%)- column

print("Discount Rate (%) :", df['Discount Rate (%)'].mean())
print("Discount Rate (%) Standard deviation :", df['Discount Rate (%)'].std())
print("Discount Rate (%) Median:", (df['Discount Rate (%)']).median())
print("Discount Rate (%) Mode:",df['Discount Rate (%)'].mode())
print("=============================================================")

#Discount Amount - Column

print("Discount Amount Mean :", df['Discount Amount'].mean())
print("Discount Amount Standard deviation :", df['Discount Amount'].std())
print("Discount Amount Median:", (df['Discount Amount']).median())
print("Discount Amount Mode:",df['Discount Amount'].mode())
print("=============================================================")



#Net Sales Value - Column

print("Net Sales Value Mean :", df['Net Sales Value'].mean())
print("Net Sales Value Standard deviation :", df['Net Sales Value'].std())
print("Net Sales Value Median:", (df['Net Sales Value']).median())
print("Net Sales Value Mode:",df['Net Sales Value'].mode())
print("=============================================================")
```

```python
# Step 4: -
# Interpretation: 
# 1.	The sample mean gives the central value of the column by averaging all the random variable values. Mean will get influenced by outliers.
# 2.	The mode is the value that occurs most frequently in the dataset. If we can’t use mean and median then mode will be used.
# 3.	The median represents the middle value when the data is ordered. It's a better indicator of central tendency than the mean when the data has outliers or is skewed. 
# 4.	Standard Deviation provides a measure of how much the data points deviate from the mean. 
```

```python
# Standardization of Numerical Variables:
```

```python
 # •	Concept: The dataset contains many independent variables, each measured in different units. This makes it difficult to predict the dependent variable accurately by combining them directly. To address this, we transform all the independent variables into a standard, unit-free form using Z-score normalization, also known as standardization.
# •	Z-score normalization transforms the original numerical variables, which may have different ranges, into a standardized range typically between -3σ and +3σ. This process is known as standardization.
# •	Formula: - Z=X−μ/ σ
```

```python
# Standardizing the data distribution:

num = df.select_dtypes(include = "number")

from sklearn.preprocessing import StandardScaler

SS = StandardScaler()
SS_num = SS.fit_transform(num)

num_new = pd.DataFrame(SS_num)
num_new.columns = list(num.iloc[:,:])
print("Before Standardization :\n", num)
print("After Standardization:\n ",(num_new))
```

```python
# Conversion of Categorical Data into Dummy Variables
```

```python
# Concept: - Converting categorical data into dummy variables (one-hot encoding) is essential for many machine learning algorithms that require numerical input. 
# This process transforms each category of a feature into a new binary column (dummy variable).
# A '1' indicates the presence of that category for a given data point, while '0' indicates its absence. 
# This prevents the algorithm from incorrectly interpreting categorical values as having ordinal relationships or magnitudes. 
# By creating distinct binary features, the model can learn the independent impact of each category on the target variable.
```

```python
# Applying one-hot encoding to the categorical columns:

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

cat = df.select_dtypes(include = "object")

enc = OneHotEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
enc.fit(df[categorical_cols])
num_cols = df.select_dtypes(exclude=['object']).columns
encoded_cols = enc.transform(df[categorical_cols]).toarray()

# Create a new DataFrame with the one-hot encoded columns
encoded_df = pd.DataFrame(encoded_cols, columns=enc.get_feature_names_out(categorical_cols))
df_final = pd.concat([df[num_cols], encoded_df], axis=1)
print(" A portion of tarnsformed data:")
df_final
```

```python
# Conclusion: -
# Key findings from the descriptive analytics and data visualizations:
# 1.	By applying descriptive statistics to the numerical data, we observed that the data is not properly distributed due to two main reasons:
# •	Most of the variables are heavily influenced by extreme values, as indicated by the mean being significantly higher than the median.
# •	Many columns have a standard deviation greater than the mean, which suggests that the data is highly spread out or exhibits high variability relative to the average value.
# 2.	By applying data visualization techniques, we observed the following distribution patterns in the dataset:
# •	Most numerical variables are widely spread, as shown in the histograms.
# •	The variables are generally positively skewed, except for the 'Discount Rate' variable, which is negatively skewed.
# •	Box plots reveal that most variables contain extreme values or outliers.
# •	Bar plots of categorical variables show that the data is evenly distributed across categories for most features, except for the 'Brand' column, which shows an imbalanced distribution.
# Importance of data preprocessing steps like standardization and one-hot encoding in data analysis and machine learning:
# •	The importance of standardization in machine learning and data analysis lies in ensuring that the data is clean, consistent, and suitable for analysis. Many algorithms are sensitive to differences in feature units or scales, which can negatively affect model performance. Standardization addresses this by transforming features with different scales into a common scale.
# For example, features like ‘Discount Rate’ and ‘Net Sales Value’ may have vastly different ranges. Standardization brings them to the same scale, allowing the model to treat them equally during training.
# •	One-hot encoding, also known as creating dummy variables, is a technique used to convert categorical variables into binary (0/1) columns—one for each category. This is important because most machine learning algorithms cannot work with text labels directly and require numerical input. One-hot encoding also helps avoid false ordinal relationships, as it treats each category independently, preventing the model from assuming that one category is greater or less than another.
```
